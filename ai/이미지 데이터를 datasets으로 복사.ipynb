# labels개수와 images개수가 같은지 확인

```python
labels = glob(os.getcwd()+"/labels/*.txt")
images = glob(os.getcwd()+"/images/*.png")
print(len(labels), len(images))
```

```python
label_set = set()

def make_filename_set(paths):
    ret = set()
    for path in paths:
        filename, _ = os.path.splitext(path)
        filename = filename.split("/")[-1]
        ret.add(filename)
    return ret

label_set = make_filename_set(labels)
image_set = make_filename_set(images)
```

```python
label_set.difference(image_set)
```
