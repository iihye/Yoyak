```python
import pickle
```

```python
with open("json_dict.pkl", "rb") as f:
    load_data = pickle.load(f)
len(load_data)
```

```python
load_data
```

```python
code2idx ={}
for idx, (key, value) in enumerate(load_data.items()):
#     print(idx, key, value)
    code2idx[key] = [idx, value]
#     print(key, code2idx[key])
```

```python
import pandas as pd
```

```python
parsed_data = []
for key, value in code2idx.items():
#     print(value)
    item_seq = str(key)
    idx = value[0]
    drug_n = value[1][0]
    name = value[1][1]
    group = value[1][2]
#     print(item_seq, idx, drug_n, name)
    parsed_data.append([item_seq, idx, drug_n, name, group])

df = pd.DataFrame(parsed_data, columns=["item_seq", "idx", "drug_n", "name", "group"])
type(df)
```

```python
df.to_csv("code2idx.csv", index=False)
```
